```python
import pandas as pd
import json

PROJECT_ROOT_DIR = "/content/drive/MyDrive/02_research/novel-gnn-ad-sh"

selected_devices_path = f"{PROJECT_ROOT_DIR}/data/CU_selected_devices.json"
data_path = "/content/drive/MyDrive/02_research/td_gdn/implementation/data/original/CUMaster.csv"

# ----------------------------
# Load selected devices config
# ----------------------------
with open(selected_devices_path, "r", encoding="utf-8") as f:
    devices_config = json.load(f)

selected_devices_names = []

for key, value in devices_config.items():
    if isinstance(value, dict) and "devices" in value:
        selected_devices_names.extend(value["devices"])

print(f"Selected devices in JSON: {len(selected_devices_names)}")

# ---------------------------------------------------------
# Read ONLY the CSV header (very memory efficient)
# ---------------------------------------------------------
csv_columns = pd.read_csv(data_path, nrows=0).columns
csv_columns = csv_columns.str.strip()

print(f"Total columns in CSV: {len(csv_columns)}")

# ---------------------------------------------------------
# Determine which requested columns actually exist
# ---------------------------------------------------------
selected_devices = [
    col for col in selected_devices_names
    if col in csv_columns
]

# Keep Timestamp column
if "Timestamp" in csv_columns and "Timestamp" not in selected_devices:
    selected_devices.append("Timestamp")

print(f"Selected columns found in CSV: {len(selected_devices)}")

# ---------------------------------------------------------
# Load ONLY required columns
# ---------------------------------------------------------
df_filtered = pd.read_csv(
    data_path,
    usecols=lambda c: c.strip() in selected_devices
)

# Remove trailing spaces from column names
df_filtered.columns = df_filtered.columns.str.strip()

print("Filtered dataframe shape:", df_filtered.shape)

# ---------------------------------------------------------
# Save result
# ---------------------------------------------------------
save_path = f"{PROJECT_ROOT_DIR}/data/CUMaster2.csv"
df_filtered.to_csv(save_path, index=False)

print(f"Saved to: {save_path}")
```
